# Load the data

Place the files in the ../../05_src/data/fires/ directory and load the appropriate file. 

```python
# Load the libraries as required.
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, FunctionTransformer
from sklearn.pipeline import Pipeline
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
import shap
from sklearn.model_selection import train_test_split
```

```python
# Load data
columns = [
    'coord_x', 'coord_y', 'month', 'day', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain', 'area' 
]
fires_dt = (pd.read_csv('../../05_src/data/fires/forestfires.csv', header = 0, names = columns))
fires_dt.info()
```

# Get X and Y

Create the features data frame and target data.

```python
X = fires_dt.drop(columns=["area"]) 
Y = fires_dt["area"] 
X_train, X_test, y_train, y_test = train_test_split(X, Y , random_state=42, test_size=0.30)
```

# Preprocessing

Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

- Numerical variables

    * (Preproc 1 and 2) Scaling: use a scaling method of your choice (Standard, Robust, Min-Max). 
    * Preproc 2 only: 
        
        + Choose a transformation for any of your input variables (or several of them). Evaluate if this transformation is convenient.
        + The choice of scaler is up to you.

- Categorical variables: 
    
    * (Preproc 1 and 2) Apply [one-hot encoding](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) where appropriate.


+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python

numerical_cols = ["coord_x", "coord_y", "ffmc", "dmc", "dc", "isi", "temp", "rh", "wind", "rain"]
categorical_cols = ["month", "day"]

numerical_transformer = StandardScaler()  
categorical_transformer = OneHotEncoder(handle_unknown="ignore")  

preproc1 = ColumnTransformer(
    transformers=[
        ("num", numerical_transformer, numerical_cols), 
        ("cat", categorical_transformer, categorical_cols)  
    ]
)

print(preproc1)
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
non_linear_cols = ["rain", "isi"] 
linear_cols = [col for col in numerical_cols if col not in non_linear_cols] 

log_transformer = Pipeline(steps=[
    ("log", FunctionTransformer(np.log1p, validate=True)), 
    ("scaler", StandardScaler()) 
])

preproc2 = ColumnTransformer(
    transformers=[
        ("log_num", log_transformer, non_linear_cols), 
        ("linear_num", numerical_transformer, linear_cols), 
        ("cat",categorical_transformer, categorical_cols) 
    ]
)

print(preproc2)
```

## Model Pipeline


Create a [model pipeline](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html): 

+ Add a step labelled `preprocessing` and assign the Column Transformer from the previous section.
+ Add a step labelled `regressor` and assign a regression model to it. 

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Choose a more advance regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Both model choices are up to you, feel free to experiment.

```python
# Pipeline A = preproc1 + baseline

baseline_regressor = KNeighborsRegressor()

pipeline1 = Pipeline(steps=[
    ("preprocessor", preproc1),
    ("regressor", baseline_regressor)
])

param_grid1 = {"regressor__n_neighbors": [3, 5, 7]}  
grid1 = GridSearchCV(pipeline1, param_grid1, cv=3, scoring="neg_mean_squared_error")  

grid1.fit(X, Y)

```

```python
# Pipeline B = preproc2 + baseline

advanced_regressor = RandomForestRegressor(random_state=42)

pipeline2 = Pipeline(steps=[
    ("preprocessor", preproc1),
    ("regressor", advanced_regressor)
])

param_grid2 = {
    "regressor__n_estimators": [50, 100],  
    "regressor__max_depth": [None, 10]
}
grid2 = GridSearchCV(pipeline2, param_grid2, cv=3, scoring="neg_mean_squared_error")

grid2.fit(X, Y)
```

```python
# Pipeline C = preproc1 + advanced model

pipeline3 = Pipeline(steps=[
    ("preprocessor", preproc2),
    ("regressor", baseline_regressor)
])


param_grid3 = {"regressor__n_neighbors": [3, 5, 7]}
grid3 = GridSearchCV(pipeline3, param_grid3, cv=3, scoring="neg_mean_squared_error")


grid3.fit(X, Y)


```

```python
# Pipeline D = preproc2 + advanced model

pipeline4 = Pipeline(steps=[
    ("preprocessor", preproc2),
    ("regressor", advanced_regressor)
])

param_grid4 = {
    "regressor__n_estimators": [50, 100],
    "regressor__max_depth": [None, 10]
}
grid4 = GridSearchCV(pipeline4, param_grid4, cv=3, scoring="neg_mean_squared_error")

grid4.fit(X, Y)
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.
+ Experiment with at least four value combinations per pipeline.

```python
param_grid1 = {
    "regressor__n_neighbors": [3, 5, 7, 9]
}

scoring_metrics = [
    "neg_mean_absolute_error", 
    "neg_mean_squared_error", 
    "neg_root_mean_squared_error"
]

grid_cv1 = GridSearchCV(
    estimator=pipeline1, 
    param_grid=param_grid1, 
    scoring=scoring_metrics, 
    cv=5, 
    refit = "neg_root_mean_squared_error"
)

grid_cv1.fit(X_train, y_train)
res1 = grid_cv1.cv_results_

```

```python
param_grid2 = {
    "regressor__n_estimators": [50, 100, 200],
    "regressor__max_depth": [None, 10, 20]
}

scoring_metrics = [
    "neg_mean_absolute_error",
    "neg_mean_squared_error",
    "neg_root_mean_squared_error"
]

grid_cv2 = GridSearchCV(
    estimator=pipeline2,
    param_grid=param_grid2,
    scoring=scoring_metrics,
    cv=5,
    refit="neg_root_mean_squared_error"
)

grid_cv2.fit(X_train, y_train)
res2 = grid_cv2.cv_results_

```

```python
param_grid3 = {
    "regressor__n_neighbors": [3, 5, 7, 9]
}

scoring_metrics = [
    "neg_mean_absolute_error",
    "neg_mean_squared_error",
    "neg_root_mean_squared_error"
]

grid_cv3 = GridSearchCV(
    estimator=pipeline3,
    param_grid=param_grid3,
    scoring=scoring_metrics,
    cv=5,
    refit="neg_root_mean_squared_error"
)

grid_cv3.fit(X_train, y_train)
res3 = grid_cv3.cv_results_

```

```python
param_grid4 = {
    "regressor__n_estimators": [50, 100, 200],
    "regressor__max_depth": [None, 10, 20]
}

scoring_metrics = [
    "neg_mean_absolute_error",
    "neg_mean_squared_error",
    "neg_root_mean_squared_error"
]

grid_cv4 = GridSearchCV(
    estimator=pipeline4,
    param_grid=param_grid4,
    scoring=scoring_metrics,
    cv=5,
    refit="neg_root_mean_squared_error"
)

grid_cv4.fit(X_train, y_train)
res4 = grid_cv4.cv_results_

```
